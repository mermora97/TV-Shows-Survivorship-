# Load data 

```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)  # Optional: Make the output less wrapped
# Display up to 100 rows of data
pd.options.display.min_rows = 100

# Load data
data = pd.read_csv('../data/my_cleaned_data.csv')
data.head()
```

```python
# First, Sort by show, then by episode (season+episode)
data = data.sort_values(['Show', 'Season', 'Episode', 'Air Date']).reset_index(drop=True)
# Second, list columns we want to create lags for
emotion_cols = [col for col in data.columns if col.startswith('avg_')]
emotion_cols

# Third, we create lagged features within each show
for col in emotion_cols:
    # For 1-lag (previous episode)
    data[f'{col}_lag'] = data.groupby('Show')[col].shift(1)
    
    # If you want multiple lags, add them in a loop
    # for lag in [1, 2, 3, ...]:
    #     data[f'{col}_lag{lag}'] = data.groupby('Show')[col].shift(lag)
```

```python
# Extract features
lagged_emotion_cols = [col for col in data.columns if col.endswith('_lag')]
model_data = data[lagged_emotion_cols + ['vw_diff', 'vw_pct_change']]

# Drop null values (first episodes)
model_data = model_data.dropna()

model_data.head()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Separate features from predictor
X = model_data[lagged_emotion_cols]
y = model_data['vw_pct_change']

# Split into train and test
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.2,        # 30% test
    shuffle=False,        # maintain chronological order
    random_state=42
)
```

```python
# Train a Random Forest Regressor
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
```

```python
import numpy as np
# 7 Evaluate on the Test Set
y_pred = rf.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print(f"Random Forest MAE:  {mae:.2f}")
print(f"Random Forest RMSE: {rmse:.2f}")
```
